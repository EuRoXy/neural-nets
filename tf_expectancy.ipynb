```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler
```

```python
import os
os.environ['TF_CPP_MIN_LOG_LEVEL'] = '2'
```

```python
import tensorflow
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import InputLayer
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam
```

```python
ds = pd.read_csv('life_expectancy.csv') # load the dataset
```

```python
ds.head()
```

```python
ds.info()
```

```python
ds['Life expectancy'].describe()
```

```python
df = ds.drop(['Country'], axis=1)
```

```python
labels = df['Life expectancy']
```

```python
features = df.iloc[:,0:-1]
features.head()
```

### data processing

```python
features = pd.get_dummies(features)
features.head()
```

```python
features_train, features_test, labels_train, labels_test = train_test_split(features, labels, 
                test_size=0.2, random_state=23)
```

```python
num_features = features.select_dtypes(include=['float64', 'int64'])
num_cols = num_features.columns
```

```python
# standardize
ct = ColumnTransformer([('only_numeric', StandardScaler(), num_cols)], remainder='passthrough')
```

```python
features_train_scaled = ct.fit_transform(features_train)
features_test_scaled = ct.transform(features_test)
```

### build the model

```python
my_model = Sequential()
input = InputLayer(shape=(features.shape[1],)) 
# add the input layer
my_model.add(input) 
# add a hidden layer with 64 neurons
my_model.add(Dense(64, activation='relu')) 
# add an output layer to our model
my_model.add(Dense(1)) 
```

```python
print(my_model.summary())
```

### Initialize optimizer and compile model

```python
opt = Adam(learning_rate=0.01)
my_model.compile(loss='mse',  metrics=['mae'], optimizer=opt)
```

### Fit and evaluate model

```python
my_model.fit(features_train_scaled, labels_train, epochs=40, 
             batch_size=1, verbose=1)
```

```python
# evaluate the model on the test data
res_mse, res_mae = my_model.evaluate(features_test_scaled, labels_test, verbose=0)
print(res_mse, res_mae)
```
